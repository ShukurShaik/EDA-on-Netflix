```python
# importing required libraries
import pandas as pd
%matplotlib inline
import numpy as np
import seaborn as sns
```

```python
f=pd.read_csv("Netflix.csv")
f.head()
```

```python
f.info()
```

```python
f["release year"].value_counts().plot.bar(figsize=(12,6))
```

```python
h=f[(f["release year"]>=2015)]
```

```python
d=h[["release year","channel","webseries","genre","user rating score"]]
```

```python
d.head(10)
```

```python
# look at the information of each year from 2015-2017 and then make inference. 
d[d["release year"]==2015].info()
d[d["release year"]==2016].info()
d[d["release year"]==2017].info()
```

```python
# Objet: find what happened in 2015 & 2016. and why the netflix ratings fall down in 2017.
# just devide the netflix data to make analysis on the year-wise for to make an inference from the data and to get the better understandig of the data and solve the problems.  
x=d[d["release year"]==2015]
y=d[d["release year"]==2016]
z=d[d["release year"]==2017]
```
